### Importing libs

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

### Load dataSet from CSV file 

```python
import pandas as pd
dataSetAddress = r"D:\Payam\UT\Machin Learning\MyCode\Clusterring\CustomerClustering\Cust_Segmentation.csv"
cus_df = pd.read_csv(dataSetAddress)
cus_df.head()
```

### Pre-processing 

```python
df = cus_df.drop("Address" , axis=1)
df.head()
```

```python
X = df.values[:,1:]
X = np.nan_to_num(X)
clus_dataSet = StandardScaler().fit_transform(X)
clus_dataSet
```

### Modelling

```python
ClusterNum = 3
K_means = KMeans(init = "k-means++" , n_clusters = ClusterNum , n_init=12)
K_means.fit(X)
labels = K_means.labels_
labels
```

```python
df["Clus_km"] = labels
df.head()
```

```python
df.groupby('Clus_km').mean()
```

```python
area = np.pi * (X[: , 1])**2
plt.scatter(X[:,0] , X[:,3] , s = area  , alpha = 0.5 , c=labels.astype(float))
plt.xlabel('Age' , fontsize=19)
plt.ylabel('Income' , fontsize=19)
```
